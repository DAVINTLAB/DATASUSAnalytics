```python
from pysus.online_data.SIH import SIH
from pathlib import Path

sih = SIH().load()

# Define a pasta de destino
pasta = Path("datasus/parquet")
pasta.mkdir(parents=True, exist_ok=True)

# Define anos e meses
anos = list(range(2008, 2023))  
meses = list(range(1, 13))      

arquivos_a_baixar = []

# Busca os arquivos para cada ano
for ano in anos:
    arquivos = sih.get_files("RD", uf="RS", year=ano, month=meses)
    arquivos_a_baixar.extend(arquivos)

# Faz o download de todos os arquivos para a pasta
baixados = sih.download(arquivos_a_baixar, local_dir=pasta)

print(f"✅ {len(baixados)} arquivos baixados com sucesso!")
```
