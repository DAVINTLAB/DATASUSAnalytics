```python
import os
import pandas as pd
import pyarrow as pa
import pyarrow.parquet as pq
from glob import glob

def unificar_parquets_chunks(pasta_parquets, saida_parquet, colunas_desejadas, batch_size=500_000):
    """
    Une arquivos Parquet de uma pasta (ignorando pastas), mantm colunas desejadas, e salva em um novo Parquet em batches.
    
    Args:
        pasta_parquets (str): Pasta com arquivos Parquet.
        saida_parquet (str): Caminho do Parquet de sada.
        colunas_desejadas (list): Lista de colunas a manter.
        batch_size (int): Tamanho dos batches para processamento.
    """

    os.makedirs(os.path.dirname(saida_parquet), exist_ok=True)

    arquivos = glob(os.path.join(pasta_parquets, "**/*.parquet"), recursive=True)
    arquivos = [f for f in arquivos if os.path.isfile(f)]

    if not arquivos:
        print("Nenhum arquivo .parquet encontrado na pasta.")
        return

    writer = None

    for arquivo in arquivos:
        print(f"Processando: {arquivo}")

        try:
            parquet_file = pq.ParquetFile(arquivo)

            for batch in parquet_file.iter_batches(batch_size=batch_size):
                df = batch.to_pandas()

                # Adiciona colunas ausentes como NaN
                for coluna in colunas_desejadas:
                    if coluna not in df.columns:
                        df[coluna] = pd.NA

                # Mantm apenas as colunas desejadas
                df = df[colunas_desejadas]

                table = pa.Table.from_pandas(df, preserve_index=False)

                if writer is None:
                    writer = pq.ParquetWriter(saida_parquet, table.schema)
                writer.write_table(table)

        except Exception as e:
            print(f" Erro ao processar {arquivo}: {e}")
            continue

    if writer:
        writer.close()
        print(f"Arquivo {os.path.basename(saida_parquet)} criado com sucesso!")


pasta = "datasus/parquet"
saida = "datasus/parquet_unificado/sih_rs.parquet"
colunas = [
    'ESPEC', 'N_AIH', 'IDENT', 'CEP', 'MUNIC_RES', 'NASC', 'SEXO', 'DT_INTER', 'DT_SAIDA',
    'UTI_MES_TO', 'MARCA_UTI', 'UTI_INT_TO', 'DIAR_ACOM', 'QT_DIARIAS', 'PROC_SOLIC', 'PROC_REA', 
    'VAL_SH', 'VAL_SP', 'VAL_TOT', 'VAL_UTI', 'NATUREZA', 'CNES', 'NAT_JUR', 'GESTAO', 'IND_VDRL', 
    'IDADE', 'DIAG_PRINC', 'DIAG_SECUN', 'COBRANCA', 'MORTE', 'MUNIC_MOV', 'DIAS_PERM', 'NACIONAL', 
    'NUM_FILHOS', 'INSTRU', 'CID_NOTIF', 'CONTRACEP1', 'CONTRACEP2', 'GESTRICO', 'INSC_PN', 'CBOR',
    'CNAER', 'VINCPREV', 'INFEHOSP', 'CID_ASSO', 'CID_MORTE', 'COMPLEX', 'RACA_COR', 'ETNIA',
    'DIAGSEC1', 'DIAGSEC2', 'DIAGSEC3', 'DIAGSEC4', 'DIAGSEC5', 'DIAGSEC6', 'DIAGSEC7', 'DIAGSEC8', 
    'DIAGSEC9', 'COD_IDADE'
]

unificar_parquets_chunks(pasta, saida, colunas)
```

```python
import pandas as pd
import numpy as np
import os
import pyarrow.parquet as pq

# Funes de tratamento
def tratar_num_filhos(df):
    if 'NUM_FILHOS' in df.columns:
        df['NUM_FILHOS'] = pd.to_numeric(df['NUM_FILHOS'], errors='coerce').fillna(0).astype(int)
        media_filhos = df.loc[df['NUM_FILHOS'] <= 20, 'NUM_FILHOS'].mean()
        df.loc[df['NUM_FILHOS'] > 20, 'NUM_FILHOS'] = media_filhos
    return df

def tratar_instrucao(df):
    if 'INSTRU' not in df.columns:
        return df
    df['INSTRU'] = df['INSTRU'].astype(str).str.zfill(2)
    df['INSTRU'] = df['INSTRU'].replace(['00', 'nan'], '0')
    return df

def padronizar_cids(df, colunas_cid):
    colunas_existentes = [col for col in colunas_cid if col in df.columns]
    if not colunas_existentes:
        return df
    print(f"Tratando {len(colunas_existentes)} variveis CID...")
    for col in colunas_existentes:
        df[col] = df[col].astype(str).str.upper().str.strip()
        df[col] = df[col].replace(['0000', 'NAN', ''], np.nan)
    return df

def tratar_cids(df):
    colunas_cid = [
        'DIAG_PRINC', 'DIAG_SECUN', 'CID_NOTIF', 'CID_ASSO', 'CID_MORTE'
    ] + [f'DIAGSEC{i}' for i in range(1, 10)]
    return padronizar_cids(df, colunas_cid)

def tratar_idade(df):
    if 'IDADE' in df.columns:
        df['IDADE'] = pd.to_numeric(df['IDADE'], errors='coerce').fillna(0).astype(int)

    if 'COD_IDADE' in df.columns:
        df['COD_IDADE'] = pd.to_numeric(df['COD_IDADE'], errors='coerce').fillna(0).astype(int)
        
        cond_horas = df['COD_IDADE'] == 1
        cond_dias = df['COD_IDADE'] == 2
        cond_meses = df['COD_IDADE'] == 3
        
        df.loc[cond_horas, 'IDADE'] = 0
        df.loc[cond_dias, 'IDADE'] = (df.loc[cond_dias, 'IDADE'] / 365).round(1)
        df.loc[cond_meses, 'IDADE'] = (df.loc[cond_meses, 'IDADE'] / 12).round(1)
    return df

def tratar_sexo(df):
    if 'SEXO' not in df.columns:
        return df
    print("Tratando coluna SEXO...")
    df['SEXO'] = pd.to_numeric(df['SEXO'], errors='coerce').astype('int64')
    return df

def tratar_datas(df):
    colunas_datas = ['DT_INTER', 'DT_SAIDA', 'NASC']
    colunas_existentes = [col for col in colunas_datas if col in df.columns]
    for col in colunas_existentes:
        df[col] = pd.to_datetime(df[col], errors='coerce', format='%Y%m%d')
        
    return df

def tratar_valores(df: pd.DataFrame) -> pd.DataFrame:
    """Converte as colunas de valores monetrios de object para float."""
    colunas_valores = ['VAL_SH', 'VAL_SP', 'VAL_TOT', 'VAL_UTI']
    
    for col in colunas_valores:
        if col in df.columns:
            print(f"Tratando coluna {col}...")
            df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0).astype(float)
    
    return df

def tratar_inteiros(df: pd.DataFrame) -> pd.DataFrame:
    """Converte as colunas inteiras de object para int."""
    colunas_inteiras = ['UTI_MES_TO', 'UTI_INT_TO', 'DIAR_ACOM', 'QT_DIARIAS', 'DIAS_PERM']
    
    for col in colunas_inteiras:
        if col in df.columns:
            print(f"Tratando coluna {col}...")
            df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0).astype(int)
    
    return df



# Funo para aplicar mltiplos tratamentos
def aplicar_tratamentos(df, tratamentos):
    for func in tratamentos:
        df = func(df)
    return df

# Processar Parquet em chunks
def processar_chunks(parquet_entrada, parquet_saida, tratamentos, batch_size=500_000):
    parquet_file = pq.ParquetFile(parquet_entrada)
    for i, batch in enumerate(parquet_file.iter_batches(batch_size=batch_size)):
        df_chunk = batch.to_pandas()
        print(f"Processando chunk {i+1}...")

        df_chunk = aplicar_tratamentos(df_chunk, tratamentos)

        df_chunk.to_parquet(
            parquet_saida,
            engine="fastparquet",
            compression="snappy",
            index=False,
            append=(i > 0)
        )
    print(f"Parquet tratado salvo em: {parquet_saida}")

# Lista de tratamentos
tratamentos = [
    tratar_num_filhos,
    tratar_instrucao,
    tratar_cids,
    tratar_idade,
    tratar_sexo,
    tratar_datas,
    tratar_valores,
    tratar_inteiros
    
    
]

# Caminhos
parquet_entrada = "datasus/parquet_unificado/sih_rs.parquet"
parquet_saida = "datasus/parquet_unificado/sih_rs_tratado.parquet"

# Executar
processar_chunks(parquet_entrada, parquet_saida, tratamentos, batch_size=500_000)
```
