```python
import pandas as pd
from sqlalchemy import create_engine

# Caminho do parquet e nome da tabela destino
parquet_path = "datasus/parquet_unificado/sih_rs_tratado.parquet"
nome_tabela = "sih_rs"

# Dados de conexão com PostgreSQL
usuario = "postgres"
senha = "1234"
host = "localhost"
porta = "5432"
banco = "sih_rs"

# Conectar ao PostgreSQL
engine = create_engine(f"postgresql+psycopg2://{usuario}:{senha}@{host}:{porta}/{banco}")

# Carrega todo o Parquet
df = pd.read_parquet(parquet_path, engine="pyarrow")  

# Envia em pedaços manualmente
chunksize = 50000
for i in range(0, len(df), chunksize):
    chunk = df.iloc[i:i+chunksize]
    chunk.to_sql(nome_tabela, engine, if_exists="append" if i > 0 else "replace", index=False)
    print(f"Chunk {i//chunksize + 1} inserido com sucesso!")
```

```python
import os
import pandas as pd
from sqlalchemy import create_engine

# Dados do banco de dados
usuario = 'postgres'
senha = '1234'
host = 'localhost'
porta = '5432'
banco = 'sih_rs'

# Conexão
engine = create_engine(f'postgresql+psycopg2://{usuario}:{senha}@{host}:{porta}/{banco}')

# Caminho dos CSVs
csvs = [
    'datasus/municipios_cod.csv',
    'datasus/cid10.csv',
    'datasus/procedimentos.csv'

]

for caminho_csv in csvs:
    nome_tabela_csv = os.path.splitext(os.path.basename(caminho_csv))[0].lower()

    print(f"Carregando CSV: {caminho_csv}...")
    df_csv = pd.read_csv(caminho_csv, sep=';', encoding='utf-8') 
    df_csv.to_sql(nome_tabela_csv, engine, if_exists='replace', index=False)
    print(f"Tabela '{nome_tabela_csv}' criada no banco!")

print("Todas as tabelas foram inseridas com sucesso no banco de dados!")
```
